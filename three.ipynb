```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=2,alpha=.5,ls='-',marker='*',ms=10,
       mfc='y', mec='g')
# lw -> line width
# alpha -> transpency
# ls -> line style
# ms -> marker size
# mfc -> marker face color
# mec -> marker edge color
#ax.set_xlim([0,1])
#ax.set_ylim([0,2])
```
